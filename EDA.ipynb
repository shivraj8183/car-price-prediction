```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
company_vehiclecounts = ds["company"].value_counts()

ax = company_vehiclecounts.plot(kind = 'bar', figsize=(12, 6), color='skyblue')

plt.title("Number of Vehicles per Company")

plt.xlabel("Company")

plt.ylabel("Vehicle Count")

plt.xticks(rotation=45)

#Add count labels on top of bars

for i, value in enumerate (company_vehiclecounts):

    plt.text(i, value + 1, str(value), ha='center', va='bottom', fontsize=9)
    plt.tight_layout()

plt.show()
```

```python
fuel_type_vechicalcount =  ds["fuel_type"].value_counts()
fuel_type_vechicalcount
ax = fuel_type_vechicalcount.plot(kind = 'bar', figsize=(12, 6), color='skyblue')

plt.title("Number of Vehicles on fuel type")

plt.xlabel("Year")

plt.ylabel("Vehicle Count")

plt.xticks(rotation=45)

#Add count labels on top of bars

for i, value in enumerate (fuel_type_vechicalcount):

    plt.text(i, value + 1, str(value), ha='center', va='bottom', fontsize=9)

plt.tight_layout()

plt.show()
```

```python
fuel_type_counts = ds["fuel_type"].value_counts()

plt.figure(figsize=(4, 4))

plt.pie(
    fuel_type_counts,
    labels=fuel_type_counts.index,
    autopct="%1.1f%%",
    startangle=90,
    colors=plt.cm.Set3.colors,   # nice color palette
    explode=[0.05]*len(fuel_type_counts),  # slight separation for all slices
)

plt.title("Distribution of Vehicles by Fuel Type", fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

sns.swarmplot(
    data=ds,
    x="fuel_type",
    y=None,
    size=6
)

plt.title("Swarm Plot of Vehicles by Fuel Type", fontsize=16)
plt.xlabel("Fuel Type", fontsize=14)
plt.ylabel("Vehicles", fontsize=14)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.swarmplot(data=ds, x='fuel_type', y='Price')

plt.title("Swarm Plot: Fuel Type vs Price")

plt.xlabel("Fuel Type")

plt.ylabel("Vehicle Price")

plt.xticks(rotation=45)

plt.tight_layout()
```
